# Combining Metrics

```python
import pandas as pd
import matplotlib.pyplot as plt
import operator
```

```python
# Chase's Submetric of Firearm Confiscation
firearm = pd.read_csv("https://data.wprdc.org/datastore/dump/e967381d-d7e9-48e3-a2a2-39262f7fa5c4")
firearmDF = pd.DataFrame(firearm)
firearmDF.dropna(inplace = True)
neighborhoods = []
for n in firearmDF["neighborhood"]:
    if n not in neighborhoods:
        neighborhoods.append(n)
count = dict()
for i in firearmDF.index:
    if firearmDF["neighborhood"][i] in count:
        count[firearmDF["neighborhood"][i]] += firearmDF["total_count"][i]
    else:
        count[firearmDF["neighborhood"][i]] = firearmDF["total_count"][i]

firearm_rankings = dict.fromkeys(neighborhoods, 0)
for key in count:
    firearm_rankings[key] = (count[key] / 309) * 100
firearm_rankings = dict( sorted(firearm_rankings.items(), key=operator.itemgetter(1),reverse=True))
```

```python
# Anjali's Submetric of Arrest Data
arrest = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f")
arrestDF = pd.DataFrame(arrest)

neighborhoods = []
for n in arrestDF["INCIDENTNEIGHBORHOOD"]:
    if n not in neighborhoods:
        neighborhoods.append(n)
count = dict()
for i in arrestDF.index:
    if arrestDF["INCIDENTNEIGHBORHOOD"][i] in count:
        count[arrestDF["INCIDENTNEIGHBORHOOD"][i]] += 1
    else:
        count[arrestDF["INCIDENTNEIGHBORHOOD"][i]] = 1
arrest_rankings = dict.fromkeys(neighborhoods, 0)
for key in count:
    arrest_rankings[key] = (count[key] / 3327) * 100
arrest_rankings = dict( sorted(arrest_rankings.items(), key=operator.itemgetter(1),reverse=True))
```

```python
# combining metrics into one
# a higher final ranking score indicates the least safe neighborhood
final_rankings = dict.fromkeys(neighborhoods, 0)
for key in final_rankings:
    if key in firearm_rankings:
        final_rankings[key] += (firearm_rankings[key] * .5)
    if key in arrest_rankings:
        final_rankings[key] += (arrest_rankings[key] * .5)
final_rankings = dict( sorted(final_rankings.items(), key=operator.itemgetter(1),reverse=True))
```

```python
# final bargraph of rankings

# creating bargraph
fig, ax = plt.subplots(figsize = (15, 10))
ax.bar(final_rankings.keys(), final_rankings.values())
ax.set_title('Ranking Neighborhoods Based on Safety')
ax.set_ylabel('Safety Scale (closer to 0 - safe) (closer to 100 - unsafe)')
# rotating labels
plt.xticks(rotation=90)
plt.show()
```

# Conclusion

Our individual analysis saw an overlap and with our metrics combined, Homewood South & Southside Flats are the least safe neighborhoods while Mt. Oliver & Regent Square were the safest
